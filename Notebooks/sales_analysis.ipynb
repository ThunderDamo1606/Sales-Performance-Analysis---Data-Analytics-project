##  Handle missing values and clean data.

```python



df["Customer_Name"] = df["Customer_Name"].fillna("Unknown")

df = df.dropna(subset=["Order_Date"])
df["Order_Date"] = pd.to_datetime(df["Order_Date"], errors="coerce")

df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")
df["Profit"] = pd.to_numeric(df["Profit"], errors="coerce")
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")

df["LineTotal"] = df["Amount"]
df["YearMonth"] = df["Order_Date"].dt.to_period("M").astype(str)



df["YearMonth"].head(10)

```

```python

df["Customer_Name"].head(10)

```

```python

df["LineTotal"].head(20)
```

## Analyze total sales per customer.

```python



total_sales_per_customer = (
    df.groupby("Customer_Name")["LineTotal"]
      .sum()
      .reset_index()
      .sort_values("LineTotal", ascending=False)
)
total_sales_per_customer.head(10)


```

## Calculate moving average sales per month.

```python




monthly = (
    df.groupby("YearMonth")["LineTotal"]
      .sum()
      .reset_index()
      .sort_values("YearMonth")
)

monthly["MovingAvg_3M"] = monthly["LineTotal"].rolling(3).mean()
monthly

```

## Segment customers by total spending.

```python




customer_total = (
    df.groupby("Customer_Name")["LineTotal"]
      .sum()
      .reset_index()
)

customer_total["Segment"] = pd.qcut(
    customer_total["LineTotal"],
    q=3,
    labels=["Low", "Medium", "High"]
)

customer_total.head(10)

```

## Calculate product return rate.

```python




df["Returned"] = df["Profit"] < 0
return_rate = df["Returned"].mean() * 100

return_rate.round(2)

```

## Identify top 10 customers by lifetime value.

```python



top10_ltv = (
    df.groupby("Customer_Name")["LineTotal"]
      .sum()
      .reset_index()
      .sort_values("LineTotal", ascending=False)
      .head(10)
)
top10_ltv

```

# Graph

## State-wise Revenue Chart

```python



state_group = df.groupby("State")["LineTotal"].sum().sort_values()

plt.figure(figsize=(12,6))
state_group.plot(kind='bar')
plt.title("Revenue by State")
plt.xlabel("State")
plt.ylabel("Revenue")
plt.grid()
plt.show()
```

## Category-wise Revenue Chart

```python


cat_group = df.groupby("Category")["LineTotal"].sum()

plt.figure(figsize=(12,6))
cat_group.plot(kind='bar')
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.grid()
plt.show()
```

## Sub-Category-wise Sales

```python


subcat_group = df.groupby("Sub_Category")["LineTotal"].sum()

plt.figure(figsize=(12,6))
subcat_group.plot(kind='bar')
plt.title("Revenue by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Revenue")
plt.grid()
plt.show()
```

## Customer-wise Revenue (Top 10)

```python


top10 = df.groupby("Customer_Name")["LineTotal"].sum().sort_values().tail(10)

plt.figure(figsize=(12,6))
top10.plot(kind='barh')
plt.title("Top 10 Customers by Revenue")
plt.xlabel("Revenue")
plt.ylabel("Customer")
plt.grid()
plt.show()
```

## City-wise Profit Chart

```python


city_profit = df.groupby("City")["Profit"].sum()

plt.figure(figsize=(12,6))
city_profit.plot(kind='bar')
plt.title("City-wise Profit")
plt.xlabel("City")
plt.ylabel("Profit")
plt.grid()
plt.show()

```

## Month-wise Revenue Trend

```python


month_group = df.groupby("YearMonth")["LineTotal"].sum()

plt.figure(figsize=(12,6))
plt.plot(month_group.index, month_group.values, marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid()
plt.show()
```

## Quantity Sold per Category

```python



qty_group = df.groupby("Category")["Quantity"].sum()

plt.figure(figsize=(12,6))
qty_group.plot(kind='bar')
plt.title("Quantity Sold per Category")
plt.xlabel("Category")
plt.ylabel("Quantity")
plt.grid()
plt.show()
```

## Profit by Category

```python


profit_group = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(12,6))
profit_group.plot(kind='bar', color='green')
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.grid()
plt.show()
```

## Daily Sales Trend

```python


daily = df.groupby(df["Order_Date"].dt.date)["LineTotal"].sum()

plt.figure(figsize=(12,6))
plt.plot(daily.index, daily.values)
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

## Weekly Sales Trend

```python


df["Week"] = df["Order_Date"].dt.isocalendar().week

weekly = df.groupby("Week")["LineTotal"].sum()

plt.figure(figsize=(12,6))
weekly.plot()
plt.title("Weekly Sales Trend")
plt.xlabel("Week")
plt.ylabel("Revenue")
plt.grid()
plt.show()
```

## Profit Margin Chart

```python



df["ProfitMargin"] = df["Profit"] / df["LineTotal"] * 100

margin_group = df.groupby("Category")["ProfitMargin"].mean()

plt.figure(figsize=(12,6))
margin_group.plot(kind='bar')
plt.title("Profit Margin by Category")
plt.xlabel("Category")
plt.ylabel("Margin (%)")
plt.grid()
plt.show()
```

## Gender-wise Customer Count

```python


gender_group = df.groupby("Gender")["Customer_Name"].count()

plt.figure(figsize=(8,6))
gender_group.plot(kind='pie', autopct="%1.1f%%", startangle=90, shadow=True)
plt.title("Customers by Gender")
plt.ylabel("")
plt.show()

```

## Sales Rep Performance Chart

```python


rep_group = df.groupby("Sales_Rep")["LineTotal"].sum()

plt.figure(figsize=(12,6))
rep_group.plot(kind='bar')
plt.title("Sales Revenue by Sales Rep")
plt.xlabel("Sales Rep")
plt.ylabel("Revenue")
plt.grid()
plt.show()
```

## Target vs Revenue Line Chart


```python


target_group = df.groupby("Month_of_Order_Date")[["LineTotal","Target"]].sum()

plt.figure(figsize=(12,6))
plt.plot(target_group.index, target_group["LineTotal"], marker='o', label="Revenue")
plt.plot(target_group.index, target_group["Target"], marker='o', label="Target")
plt.title("Target vs Revenue")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.legend()
plt.grid()
plt.show()
```
